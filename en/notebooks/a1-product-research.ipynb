## 2. Simulated Competitor Review Data

In practice, you can manually copy reviews from the Amazon storefront, or export them using tools like Helium 10.

```python
# Simulated 50 competitor reviews for a portable power bank
reviews = [
    {'rating': 5, 'text': 'Perfect for camping trips. Charged my phone 3 times on a single charge. Lightweight and easy to carry.'},
    {'rating': 4, 'text': 'Good capacity but a bit heavy. The LED indicator is useful. Wish it had USB-C input.'},
    {'rating': 1, 'text': 'Stopped working after 2 weeks. Customer service was unhelpful. Total waste of money.'},
    {'rating': 5, 'text': 'Best power bank I have owned. Fast charging works great with my iPhone. Compact design fits in pocket.'},
    {'rating': 3, 'text': 'Average product. Takes too long to recharge the power bank itself. About 6 hours for full charge.'},
    {'rating': 2, 'text': 'The actual capacity is much less than advertised. Only charged my phone 1.5 times, not 3 as claimed.'},
    {'rating': 5, 'text': 'Excellent build quality. Survived a drop on concrete. The rubberized finish feels premium.'},
    {'rating': 4, 'text': 'Great for travel. TSA approved for flights. Charged my iPad and phone simultaneously.'},
    {'rating': 1, 'text': 'Battery swelled up after 1 month. Dangerous product. Returned immediately.'},
    {'rating': 5, 'text': 'Solar charging feature is a game changer for hiking. Slow but works as backup.'},
    {'rating': 3, 'text': 'Decent but the cable is too short. Had to buy a separate longer cable.'},
    {'rating': 4, 'text': 'Good value for the price. Not the fastest charging but reliable. Using it daily for 6 months.'},
    {'rating': 2, 'text': 'The power button is too sensitive. Keeps turning on in my bag and draining battery.'},
    {'rating': 5, 'text': 'Waterproof design saved it when I dropped it in a puddle. Still works perfectly.'},
    {'rating': 4, 'text': 'Solid product. The built-in cables are convenient but wish they were longer.'},
    {'rating': 1, 'text': 'Fake reviews on this product. Real capacity is half of what is listed. Scam.'},
    {'rating': 5, 'text': 'Pass-through charging works great. Can charge the power bank and my phone at the same time.'},
    {'rating': 3, 'text': 'OK product but gets very hot during fast charging. A bit concerning.'},
    {'rating': 4, 'text': 'Nice compact size. Fits easily in my jacket pocket. Good for daily commute.'},
    {'rating': 2, 'text': 'The USB-C port stopped working after 3 months. Only the USB-A port works now.'},
]

df = pd.DataFrame(reviews)
print(f'Total {len(df)} reviews')
print(f'Rating distribution:\n{df["rating"].value_counts().sort_index()}')
```

## 3. Basic Data Analysis (Python)

```python
# Rating statistics
avg_rating = df['rating'].mean()
positive = len(df[df['rating'] >= 4])
negative = len(df[df['rating'] <= 2])
neutral = len(df[df['rating'] == 3])

print(f'Average rating: {avg_rating:.1f}/5')
print(f'Positive (4-5 stars): {positive} ({positive/len(df)*100:.0f}%)')
print(f'Negative (1-2 stars): {negative} ({negative/len(df)*100:.0f}%)')
print(f'Neutral (3 stars): {neutral} ({neutral/len(df)*100:.0f}%)')
```

```python
# Keyword frequency analysis
all_text = ' '.join(df['text'].str.lower())
# Extract meaningful words
keywords = re.findall(r'\b[a-z]{4,}\b', all_text)
# Filter stopwords
stopwords = {'this', 'that', 'with', 'from', 'have', 'been', 'were', 'they', 'their', 'about', 'would', 'could', 'after', 'very', 'much', 'than', 'also', 'just', 'only'}
keywords = [w for w in keywords if w not in stopwords]

print('Top 20 high-frequency keywords:')
for word, count in Counter(keywords).most_common(20):
    print(f'  {word}: {count}')
```

## 4. AI Deep Analysis (Prompt Template)

Copy the following Prompt + review data into ChatGPT/Claude to get a structured pain point analysis report.

```python
# Generate a Prompt ready to paste into AI
negative_reviews = df[df['rating'] <= 3]['text'].tolist()
positive_reviews = df[df['rating'] >= 4]['text'].tolist()

prompt = f"""You are a cross-border e-commerce product research analyst. Please analyze the following competitor (portable power bank) review data.

=== Negative and Neutral Reviews ({len(negative_reviews)} total) ===
{chr(10).join(f'- {r}' for r in negative_reviews)}

=== Positive Reviews ({len(positive_reviews)} total) ===
{chr(10).join(f'- {r}' for r in positive_reviews)}

Please output:
1. Pain point ranking (sorted by frequency, with occurrence count and severity for each)
2. Positive highlights (Top 5 features users love most)
3. Product selection advice (if I were to create a competing product, which 3 pain points should I prioritize solving)
4. Listing suggestions (based on positive highlights, which selling points should the listing emphasize)
"""

print('=== Copy the following into ChatGPT/Claude ===')
print(prompt)
```
