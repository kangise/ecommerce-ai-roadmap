## 1. Setup

```python
import pandas as pd
import numpy as np

print('Environment ready ✅')
```

## 2. Define Your AI Tool Investments

Modify the data below to match your actual situation.

```python
# === Modify this section with your actual data ===

ai_tools = {
    'ChatGPT Plus': {
        'monthly_cost': 20,  # USD/month
        'tasks': [
            {'name': 'Listing Generation', 'hours_saved_per_month': 8, 'hourly_rate': 25},
            {'name': 'Ad Copywriting', 'hours_saved_per_month': 4, 'hourly_rate': 25},
            {'name': 'Customer Service Templates', 'hours_saved_per_month': 3, 'hourly_rate': 20},
            {'name': 'Data Analysis', 'hours_saved_per_month': 5, 'hourly_rate': 30},
            {'name': 'Social Media Copywriting', 'hours_saved_per_month': 6, 'hourly_rate': 25},
        ]
    },
    'Helium 10': {
        'monthly_cost': 79,
        'tasks': [
            {'name': 'Keyword Research', 'hours_saved_per_month': 10, 'hourly_rate': 30},
            {'name': 'Competitor Analysis', 'hours_saved_per_month': 6, 'hourly_rate': 30},
            {'name': 'Listing Audit', 'hours_saved_per_month': 3, 'hourly_rate': 25},
        ]
    },
    'Canva Pro': {
        'monthly_cost': 13,
        'tasks': [
            {'name': 'Product Image Design', 'hours_saved_per_month': 5, 'hourly_rate': 35},
            {'name': 'Social Media Graphics', 'hours_saved_per_month': 4, 'hourly_rate': 25},
            {'name': 'Ad Creatives', 'hours_saved_per_month': 3, 'hourly_rate': 30},
        ]
    }
}

print(f'{len(ai_tools)} AI tools configured')
for name, tool in ai_tools.items():
    print(f'  - {name}: ${tool["monthly_cost"]}/month, {len(tool["tasks"])} use cases')
```

## 3. Calculate ROI for Each Tool

```python
results = []

for tool_name, tool in ai_tools.items():
    total_hours_saved = 0
    total_value_saved = 0
    task_details = []
    
    for task in tool['tasks']:
        value = task['hours_saved_per_month'] * task['hourly_rate']
        total_hours_saved += task['hours_saved_per_month']
        total_value_saved += value
        task_details.append({
            'Task': task['name'],
            'Hours Saved/Month': task['hours_saved_per_month'],
            'Hourly Rate': f"${task['hourly_rate']}",
            'Value Saved/Month': f"${value}"
        })
    
    monthly_cost = tool['monthly_cost']
    monthly_roi = ((total_value_saved - monthly_cost) / monthly_cost) * 100
    annual_net = (total_value_saved - monthly_cost) * 12
    
    results.append({
        'Tool': tool_name,
        'Monthly Cost': monthly_cost,
        'Hours Saved/Month': total_hours_saved,
        'Value Saved/Month': total_value_saved,
        'Monthly Net': total_value_saved - monthly_cost,
        'Monthly ROI': monthly_roi,
        'Annual Net Savings': annual_net
    })
    
    print(f'\n=== {tool_name} ===')
    print(pd.DataFrame(task_details).to_string(index=False))
    print(f'\nMonthly cost: ${monthly_cost} | Monthly value saved: ${total_value_saved} | Monthly net gain: ${total_value_saved - monthly_cost}')
    print(f'Monthly ROI: {monthly_roi:.0f}% | Annual net savings: ${annual_net:,.0f}')
```

## 4. Summary Comparison

```python
summary = pd.DataFrame(results)
summary['Monthly Cost'] = summary['Monthly Cost'].apply(lambda x: f'${x}')
summary['Value Saved/Month'] = summary['Value Saved/Month'].apply(lambda x: f'${x}')
summary['Monthly Net'] = summary['Monthly Net'].apply(lambda x: f'${x}')
summary['Monthly ROI'] = summary['Monthly ROI'].apply(lambda x: f'{x:.0f}%')
summary['Annual Net Savings'] = summary['Annual Net Savings'].apply(lambda x: f'${x:,.0f}')

print('=== AI Tool ROI Summary ===')
print(summary.to_string(index=False))

total_monthly_cost = sum(t['monthly_cost'] for t in ai_tools.values())
total_monthly_value = sum(r['Value Saved/Month'] for r in [{'Value Saved/Month': sum(t['hours_saved_per_month'] * t['hourly_rate'] for t in tool['tasks'])} for tool in ai_tools.values()])

print(f'\n--- Total ---')
print(f'Total monthly AI tool cost: ${total_monthly_cost}')
total_val = sum(sum(t['hours_saved_per_month'] * t['hourly_rate'] for t in tool['tasks']) for tool in ai_tools.values())
print(f'Total monthly value saved: ${total_val}')
print(f'Total monthly net gain: ${total_val - total_monthly_cost}')
print(f'Total annual net savings: ${(total_val - total_monthly_cost) * 12:,.0f}')
print(f'Overall ROI: {((total_val - total_monthly_cost) / total_monthly_cost) * 100:.0f}%')
```

## 5. Generate AI Analysis Prompt

Compile your ROI data into a prompt and let AI help you optimize your tool portfolio.

```python
tool_summary = '\n'.join([
    f"- {r['Tool']}: Monthly cost {r['Monthly Cost']}, Monthly savings {r['Value Saved/Month']}, ROI {r['Monthly ROI']}"
    for r in results
])

prompt = f"""You are an AI tool investment return analysis expert.

Below is the ROI data for the AI tools I currently use:

{tool_summary}

Total monthly cost: ${total_monthly_cost}
Total monthly savings: ${total_val}
Overall ROI: {((total_val - total_monthly_cost) / total_monthly_cost) * 100:.0f}%

Please analyze:
1. Which tool has the highest ROI? Why?
2. Are there any tools that could be replaced or canceled?
3. What other AI tools are worth considering? (Based on my use cases)
4. How can I further improve the efficiency of my AI tool usage?
5. AI tool budget recommendations for the team (monthly/annual)
"""

print('=== Copy the following prompt to ChatGPT/Claude ===')
print(prompt)
```
